## Vector

Vector is a number with a direction and have many other number that represent an value of feature or etc

```python
# Vector kumpulan nilai dalam 1d yang linear atau kalau di python list
vector = torch.tensor([1,2,3])
vector
```

## TENSOR

Tensor is an N dimensional array of numbers all of the term before is also can be called as tensor. <br>
EX: <br>


*   Scalar is 0-dim of tensor
*   Vector is 1-dim of tensor
*   Matrix is 2-dim of tensor



```python
TENSOR = torch.Tensor([[ [1,2,3],
                        [4,5,6],
                        [7,8,9] ]])

TENSOR.ndim, TENSOR.shape
```
